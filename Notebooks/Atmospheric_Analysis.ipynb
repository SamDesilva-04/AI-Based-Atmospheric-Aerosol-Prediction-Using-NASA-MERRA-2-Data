```python
# =========================================================
# FINAL TRAINING CELL
# 5 input maps -> 1 next map
# 30 epochs
# GPU
# validation + best model save + plots
# recursive weekly/monthly forecast
# =========================================================

import os, glob, re, random, math
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

# -----------------------------
# SETTINGS
# -----------------------------
DATA_FOLDER = "/content/drive/MyDrive/Colab Notebooks/NASA_Atmospheric_Project/raw_data/merra2"
SEQ_LEN = 5
BATCH_SIZE = 8
EPOCHS = 30
LR = 1e-3
WEIGHT_DECAY = 1e-5
PATIENCE = 6
NUM_WORKERS = 2
FORECAST_WEEK = 7
FORECAST_MONTH = 30

MODEL_SAVE_PATH = "/content/drive/MyDrive/Colab Notebooks/best_aerosol_cnn.pth"
WEEKLY_SAVE_PATH = "/content/drive/MyDrive/Colab Notebooks/weekly_avg_forecast.npy"
MONTHLY_SAVE_PATH = "/content/drive/MyDrive/Colab Notebooks/monthly_avg_forecast.npy"

# -----------------------------
# REPRODUCIBILITY
# -----------------------------
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
torch.cuda.manual_seed_all(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# -----------------------------
# SORT FILES BY DATE
# -----------------------------
files = glob.glob(os.path.join(DATA_FOLDER, "*.nc4"))

def extract_date(fp):
    name = os.path.basename(fp)
    m = re.search(r'(\d{8})', name)
    return m.group(1) if m else "00000000"

files = sorted(files, key=extract_date)

print("Total files found:", len(files))
print("First 3:", [os.path.basename(f) for f in files[:3]])
print("Last 3:", [os.path.basename(f) for f in files[-3:]])

if len(files) < SEQ_LEN + 50:
    raise ValueError("Not enough files for training and forecasting.")

# -----------------------------
# LOAD ALL MAPS
# each file -> mean over time -> [H, W, 2]
# -----------------------------
all_maps = []

for i, file in enumerate(files):
    try:
        ds = xr.open_dataset(file)

        bc = ds["BCSMASS"].mean("time").values.astype(np.float32)
        du = ds["DUSMASS25"].mean("time").values.astype(np.float32)

        # clean numeric issues
        bc = np.nan_to_num(bc, nan=0.0, posinf=0.0, neginf=0.0)
        du = np.nan_to_num(du, nan=0.0, posinf=0.0, neginf=0.0)

        # log transform
        bc = np.log1p(bc * 1e9)
        du = np.log1p(du * 1e9)

        stacked = np.stack([bc, du], axis=-1)  # [H, W, 2]
        all_maps.append(stacked)

        ds.close()

        if (i + 1) % 200 == 0:
            print(f"Loaded {i+1}/{len(files)} files")

    except Exception as e:
        print(f"Skipping {os.path.basename(file)} بسبب error:", e)

data = np.array(all_maps, dtype=np.float32)  # [T, H, W, 2]
print("Final data shape:", data.shape)

T, H, W, C = data.shape
assert C == 2, "Expected 2 channels"

# -----------------------------
# BUILD CHRONOLOGICAL SPLITS
# target index t uses [t-SEQ_LEN ... t-1] -> predict t
# -----------------------------
all_target_indices = np.arange(SEQ_LEN, T)

train_end = int(0.70 * len(all_target_indices))
val_end = int(0.85 * len(all_target_indices))

train_targets = all_target_indices[:train_end]
val_targets = all_target_indices[train_end:val_end]
test_targets = all_target_indices[val_end:]

print("Train samples:", len(train_targets))
print("Val samples:", len(val_targets))
print("Test samples:", len(test_targets))

# -----------------------------
# DATASET
# -----------------------------
class AerosolDataset(Dataset):
    def __init__(self, full_data, target_indices, seq_len):
        self.full_data = full_data
        self.target_indices = target_indices
        self.seq_len = seq_len

    def __len__(self):
        return len(self.target_indices)

    def __getitem__(self, idx):
        t = self.target_indices[idx]

        x = self.full_data[t-self.seq_len:t]      # [seq, H, W, 2]
        y = self.full_data[t]                     # [H, W, 2]

        x = np.transpose(x, (0, 3, 1, 2))        # [seq, 2, H, W]
        x = x.reshape(self.seq_len * 2, H, W)    # [10, H, W]

        y = np.transpose(y, (2, 0, 1))           # [2, H, W]

        return torch.tensor(x, dtype=torch.float32), torch.tensor(y, dtype=torch.float32)

train_ds = AerosolDataset(data, train_targets, SEQ_LEN)
val_ds = AerosolDataset(data, val_targets, SEQ_LEN)
test_ds = AerosolDataset(data, test_targets, SEQ_LEN)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

# -----------------------------
# BETTER CNN
# -----------------------------
class BetterAerosolCNN(nn.Module):
    def __init__(self, in_channels=10, out_channels=2):
        super().__init__()

        self.net = nn.Sequential(
            nn.Conv2d(in_channels, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Dropout(0.10),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Dropout(0.10),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Dropout(0.15),

            nn.Conv2d(128, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),

            nn.Conv2d(64, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),

            nn.Conv2d(32, out_channels, 1)
        )

    def forward(self, x):
        return self.net(x)

model = BetterAerosolCNN(in_channels=SEQ_LEN*2, out_channels=2).to(device)

criterion = nn.SmoothL1Loss()
optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode="min", factor=0.5, patience=2
)

# -----------------------------
# TRAIN
# -----------------------------
best_val_loss = float("inf")
wait = 0
train_losses = []
val_losses = []

for epoch in range(EPOCHS):
    model.train()
    running_train = 0.0

    for xb, yb in train_loader:
        xb = xb.to(device, non_blocking=True)
        yb = yb.to(device, non_blocking=True)

        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()

        running_train += loss.item()

    avg_train = running_train / max(len(train_loader), 1)
    train_losses.append(avg_train)

    model.eval()
    running_val = 0.0

    with torch.no_grad():
        for xb, yb in val_loader:
            xb = xb.to(device, non_blocking=True)
            yb = yb.to(device, non_blocking=True)

            preds = model(xb)
            loss = criterion(preds, yb)
            running_val += loss.item()

    avg_val = running_val / max(len(val_loader), 1)
    val_losses.append(avg_val)

    scheduler.step(avg_val)
    current_lr = optimizer.param_groups[0]["lr"]

    print(f"Epoch [{epoch+1}/{EPOCHS}] Train: {avg_train:.6f} | Val: {avg_val:.6f} | LR: {current_lr:.6e}")

    if avg_val < best_val_loss:
        best_val_loss = avg_val
        wait = 0
        torch.save(model.state_dict(), MODEL_SAVE_PATH)
        print("Best model saved.")
    else:
        wait += 1
        if wait >= PATIENCE:
            print("Early stopping triggered.")
            break

# -----------------------------
# LOSS PLOT
# -----------------------------
plt.figure(figsize=(8,5))
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Val Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------
# TEST
# -----------------------------
model.load_state_dict(torch.load(MODEL_SAVE_PATH, map_location=device))
model.eval()

test_loss = 0.0
with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device, non_blocking=True)
        yb = yb.to(device, non_blocking=True)
        preds = model(xb)
        loss = criterion(preds, yb)
        test_loss += loss.item()

test_loss /= max(len(test_loader), 1)
print("Final Test Loss:", test_loss)

# -----------------------------
# SHOW ONE TEST PREDICTION
# -----------------------------
xb, yb = next(iter(test_loader))
xb_gpu = xb.to(device)

with torch.no_grad():
    pred = model(xb_gpu).cpu().numpy()

true = yb.numpy()

sample_id = 0
pred_bc = pred[sample_id, 0]
pred_du = pred[sample_id, 1]
true_bc = true[sample_id, 0]
true_du = true[sample_id, 1]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0,0].imshow(true_bc)
axes[0,0].set_title("True BCSMASS")
axes[0,0].axis("off")

axes[0,1].imshow(pred_bc)
axes[0,1].set_title("Predicted BCSMASS")
axes[0,1].axis("off")

axes[1,0].imshow(true_du)
axes[1,0].set_title("True DUSMASS25")
axes[1,0].axis("off")

axes[1,1].imshow(pred_du)
axes[1,1].set_title("Predicted DUSMASS25")
axes[1,1].axis("off")

plt.tight_layout()
plt.show()

# -----------------------------
# RECURSIVE FUTURE FORECAST
# last 5 observed maps -> predict next 30 maps
# -----------------------------
history = data[-SEQ_LEN:].copy()     # [5, H, W, 2]
future_preds = []

model.eval()
with torch.no_grad():
    for step in range(FORECAST_MONTH):
        x = np.transpose(history, (0, 3, 1, 2)).reshape(1, SEQ_LEN*2, H, W)
        x = torch.tensor(x, dtype=torch.float32).to(device)

        pred = model(x).cpu().numpy()[0]       # [2, H, W]
        pred_hwc = np.transpose(pred, (1, 2, 0))   # [H, W, 2]

        future_preds.append(pred_hwc)
        history = np.concatenate([history[1:], pred_hwc[None, ...]], axis=0)

future_preds = np.array(future_preds, dtype=np.float32)
print("Future prediction shape:", future_preds.shape)

weekly_avg = future_preds[:FORECAST_WEEK].mean(axis=0)     # [H, W, 2]
monthly_avg = future_preds[:FORECAST_MONTH].mean(axis=0)   # [H, W, 2]

np.save(WEEKLY_SAVE_PATH, weekly_avg)
np.save(MONTHLY_SAVE_PATH, monthly_avg)

print("Saved weekly average forecast:", WEEKLY_SAVE_PATH)
print("Saved monthly average forecast:", MONTHLY_SAVE_PATH)

# -----------------------------
# PLOT WEEKLY + MONTHLY
# -----------------------------
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0,0].imshow(weekly_avg[:, :, 0])
axes[0,0].set_title("Weekly Avg Forecast - BCSMASS")
axes[0,0].axis("off")

axes[0,1].imshow(weekly_avg[:, :, 1])
axes[0,1].set_title("Weekly Avg Forecast - DUSMASS25")
axes[0,1].axis("off")

axes[1,0].imshow(monthly_avg[:, :, 0])
axes[1,0].set_title("Monthly Avg Forecast - BCSMASS")
axes[1,0].axis("off")

axes[1,1].imshow(monthly_avg[:, :, 1])
axes[1,1].set_title("Monthly Avg Forecast - DUSMASS25")
axes[1,1].axis("off")

plt.tight_layout()
plt.show()

print("Done.")
```

```python
# =========================================================
# REPORT GENERATION CELL
# Creates images + metrics + text summary for report
# =========================================================

import os
import numpy as np
import matplotlib.pyplot as plt
import torch

REPORT_DIR = "/content/drive/MyDrive/Colab Notebooks/atmosphere_report_outputs"
os.makedirs(REPORT_DIR, exist_ok=True)

print("Saving report files to:", REPORT_DIR)

# -----------------------------
# Helper metrics
# -----------------------------
def regression_metrics(y_true, y_pred):
    y_true = y_true.reshape(-1)
    y_pred = y_pred.reshape(-1)

    mae = np.mean(np.abs(y_true - y_pred))
    mse = np.mean((y_true - y_pred) ** 2)
    rmse = np.sqrt(mse)

    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    r2 = 1 - (ss_res / ss_tot) if ss_tot != 0 else np.nan

    mape_mask = np.abs(y_true) > 1e-8
    if np.any(mape_mask):
        mape = np.mean(np.abs((y_true[mape_mask] - y_pred[mape_mask]) / y_true[mape_mask])) * 100
    else:
        mape = np.nan

    return {
        "MAE": float(mae),
        "MSE": float(mse),
        "RMSE": float(rmse),
        "R2": float(r2),
        "MAPE_percent": float(mape)
    }

# -----------------------------
# Evaluate on entire test set
# -----------------------------
model.eval()

all_preds = []
all_true = []

with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device, non_blocking=True)
        preds = model(xb).cpu().numpy()
        all_preds.append(preds)
        all_true.append(yb.numpy())

all_preds = np.concatenate(all_preds, axis=0)   # [N, 2, H, W]
all_true = np.concatenate(all_true, axis=0)     # [N, 2, H, W]

print("Pred shape:", all_preds.shape)
print("True shape:", all_true.shape)

# -----------------------------
# Overall metrics
# -----------------------------
overall = regression_metrics(all_true, all_preds)
bc_metrics = regression_metrics(all_true[:, 0], all_preds[:, 0])
du_metrics = regression_metrics(all_true[:, 1], all_preds[:, 1])

# Per-sample MAE for extra interpretation
sample_mae = np.mean(np.abs(all_true - all_preds), axis=(1,2,3))
best_idx = int(np.argmin(sample_mae))
worst_idx = int(np.argmax(sample_mae))
median_idx = int(np.argsort(sample_mae)[len(sample_mae)//2])

print("Overall metrics:", overall)
print("BCSMASS metrics:", bc_metrics)
print("DUSMASS25 metrics:", du_metrics)
print("Best sample index:", best_idx)
print("Median sample index:", median_idx)
print("Worst sample index:", worst_idx)

# -----------------------------
# Save metrics text
# -----------------------------
summary_txt = os.path.join(REPORT_DIR, "metrics_summary.txt")
with open(summary_txt, "w") as f:
    f.write("ATMOSPHERIC AEROSOL PREDICTION - REPORT SUMMARY\n")
    f.write("="*60 + "\n\n")

    f.write("Dataset Summary\n")
    f.write(f"Total loaded maps: {data.shape[0]}\n")
    f.write(f"Spatial shape: {data.shape[1]} x {data.shape[2]}\n")
    f.write(f"Channels: {data.shape[3]}\n")
    f.write(f"Train loss final: {train_losses[-1]:.6f}\n")
    f.write(f"Validation loss final: {val_losses[-1]:.6f}\n\n")

    f.write("Overall Test Metrics\n")
    for k, v in overall.items():
        f.write(f"{k}: {v:.6f}\n")
    f.write("\n")

    f.write("BCSMASS Test Metrics\n")
    for k, v in bc_metrics.items():
        f.write(f"{k}: {v:.6f}\n")
    f.write("\n")

    f.write("DUSMASS25 Test Metrics\n")
    for k, v in du_metrics.items():
        f.write(f"{k}: {v:.6f}\n")
    f.write("\n")

    f.write("Interpretation\n")
    f.write("- Lower MAE/RMSE indicates better prediction quality.\n")
    f.write("- R2 closer to 1 indicates stronger agreement with ground truth.\n")
    f.write("- Recursive long-range forecasts may appear smoother than true maps.\n")

print("Saved:", summary_txt)

# -----------------------------
# 1. Loss curve
# -----------------------------
plt.figure(figsize=(8,5))
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, "01_loss_curve.png"), dpi=300, bbox_inches="tight")
plt.show()

# -----------------------------
# 2. Best / median / worst samples
# -----------------------------
def save_comparison_figure(idx, tag):
    true_bc = all_true[idx, 0]
    pred_bc = all_preds[idx, 0]
    res_bc  = np.abs(true_bc - pred_bc)

    true_du = all_true[idx, 1]
    pred_du = all_preds[idx, 1]
    res_du  = np.abs(true_du - pred_du)

    fig, axes = plt.subplots(2, 3, figsize=(14, 9))

    axes[0,0].imshow(true_bc)
    axes[0,0].set_title("True BCSMASS")
    axes[0,1].imshow(pred_bc)
    axes[0,1].set_title("Predicted BCSMASS")
    axes[0,2].imshow(res_bc)
    axes[0,2].set_title("Residual BCSMASS")

    axes[1,0].imshow(true_du)
    axes[1,0].set_title("True DUSMASS25")
    axes[1,1].imshow(pred_du)
    axes[1,1].set_title("Predicted DUSMASS25")
    axes[1,2].imshow(res_du)
    axes[1,2].set_title("Residual DUSMASS25")

    for ax in axes.ravel():
        ax.axis("off")

    plt.suptitle(f"{tag.capitalize()} Prediction Example (Sample {idx})")
    plt.tight_layout()
    save_path = os.path.join(REPORT_DIR, f"02_{tag}_comparison.png")
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", save_path)

save_comparison_figure(best_idx, "best")
save_comparison_figure(median_idx, "median")
save_comparison_figure(worst_idx, "worst")

# -----------------------------
# 3. Scatter plots
# -----------------------------
def save_scatter(y_true, y_pred, name):
    y_true = y_true.reshape(-1)
    y_pred = y_pred.reshape(-1)

    # sample for plotting so image stays manageable
    n = min(25000, len(y_true))
    idx = np.random.choice(len(y_true), n, replace=False)

    plt.figure(figsize=(6,6))
    plt.scatter(y_true[idx], y_pred[idx], s=3, alpha=0.3)
    mn = min(y_true[idx].min(), y_pred[idx].min())
    mx = max(y_true[idx].max(), y_pred[idx].max())
    plt.plot([mn, mx], [mn, mx], 'r--')
    plt.xlabel("Ground Truth")
    plt.ylabel("Prediction")
    plt.title(f"{name} - Prediction vs Ground Truth")
    plt.tight_layout()
    save_path = os.path.join(REPORT_DIR, f"03_scatter_{name.lower()}.png")
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", save_path)

save_scatter(all_true[:, 0], all_preds[:, 0], "BCSMASS")
save_scatter(all_true[:, 1], all_preds[:, 1], "DUSMASS25")

# -----------------------------
# 4. Histograms
# -----------------------------
def save_hist(y_true, y_pred, name):
    plt.figure(figsize=(8,5))
    plt.hist(y_true.reshape(-1), bins=60, alpha=0.6, label="Ground Truth")
    plt.hist(y_pred.reshape(-1), bins=60, alpha=0.6, label="Prediction")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title(f"{name} Value Distribution")
    plt.legend()
    plt.tight_layout()
    save_path = os.path.join(REPORT_DIR, f"04_hist_{name.lower()}.png")
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", save_path)

save_hist(all_true[:, 0], all_preds[:, 0], "BCSMASS")
save_hist(all_true[:, 1], all_preds[:, 1], "DUSMASS25")

# -----------------------------
# 5. Residual histograms
# -----------------------------
def save_residual_hist(y_true, y_pred, name):
    residual = (y_true - y_pred).reshape(-1)
    plt.figure(figsize=(8,5))
    plt.hist(residual, bins=60, alpha=0.8)
    plt.xlabel("Residual (True - Predicted)")
    plt.ylabel("Frequency")
    plt.title(f"{name} Residual Distribution")
    plt.tight_layout()
    save_path = os.path.join(REPORT_DIR, f"05_residual_hist_{name.lower()}.png")
    plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", save_path)

save_residual_hist(all_true[:, 0], all_preds[:, 0], "BCSMASS")
save_residual_hist(all_true[:, 1], all_preds[:, 1], "DUSMASS25")

# -----------------------------
# 6. Mean true vs mean predicted maps
# -----------------------------
mean_true_bc = all_true[:, 0].mean(axis=0)
mean_pred_bc = all_preds[:, 0].mean(axis=0)
mean_true_du = all_true[:, 1].mean(axis=0)
mean_pred_du = all_preds[:, 1].mean(axis=0)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0,0].imshow(mean_true_bc)
axes[0,0].set_title("Mean True BCSMASS")
axes[0,1].imshow(mean_pred_bc)
axes[0,1].set_title("Mean Predicted BCSMASS")
axes[1,0].imshow(mean_true_du)
axes[1,0].set_title("Mean True DUSMASS25")
axes[1,1].imshow(mean_pred_du)
axes[1,1].set_title("Mean Predicted DUSMASS25")
for ax in axes.ravel():
    ax.axis("off")
plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, "06_mean_maps.png"), dpi=300, bbox_inches="tight")
plt.show()

# -----------------------------
# 7. Forecast images
# -----------------------------
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0,0].imshow(weekly_avg[:, :, 0])
axes[0,0].set_title("Weekly Forecast - BCSMASS")
axes[0,1].imshow(weekly_avg[:, :, 1])
axes[0,1].set_title("Weekly Forecast - DUSMASS25")
axes[1,0].imshow(monthly_avg[:, :, 0])
axes[1,0].set_title("Monthly Forecast - BCSMASS")
axes[1,1].imshow(monthly_avg[:, :, 1])
axes[1,1].set_title("Monthly Forecast - DUSMASS25")
for ax in axes.ravel():
    ax.axis("off")
plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, "07_weekly_monthly_forecasts.png"), dpi=300, bbox_inches="tight")
plt.show()

# -----------------------------
# 8. Last observed vs monthly forecast
# -----------------------------
last_observed = data[-1]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0,0].imshow(last_observed[:, :, 0])
axes[0,0].set_title("Last Observed BCSMASS")
axes[0,1].imshow(monthly_avg[:, :, 0])
axes[0,1].set_title("Monthly Forecast BCSMASS")
axes[1,0].imshow(last_observed[:, :, 1])
axes[1,0].set_title("Last Observed DUSMASS25")
axes[1,1].imshow(monthly_avg[:, :, 1])
axes[1,1].set_title("Monthly Forecast DUSMASS25")
for ax in axes.ravel():
    ax.axis("off")
plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, "08_last_observed_vs_monthly.png"), dpi=300, bbox_inches="tight")
plt.show()

# -----------------------------
# 9. Save raw arrays for report reuse
# -----------------------------
np.save(os.path.join(REPORT_DIR, "all_test_predictions.npy"), all_preds)
np.save(os.path.join(REPORT_DIR, "all_test_ground_truth.npy"), all_true)
np.save(os.path.join(REPORT_DIR, "sample_mae.npy"), sample_mae)

print("\nReport generation complete.")
print("Main folder:", REPORT_DIR)
print("Files created:")
for f in sorted(os.listdir(REPORT_DIR)):
    print("-", f)
```
